```python
arr_c1 = [ 0.8201325456916407, 0.8283559962942604, 0.8335122025485893, 0.8423121352329465, 0.8355459449489214 ]
arr_c2 = [ 0.9371462251438739, 0.943404596342198, 0.947111895609738, 0.9510839260553026, 0.9474802947686796]
arr_c3 = [ 0.9773235976969257, 0.9806527501303584 , 0.9829315136153538, 0.9845835497049897, 0.9831857078350213]
```

```python
neg_samples = [3,6,9,12,15]
plt.figure(figsize=[6,4])
plt.xlim(2,16)
plt.ylim(0.8,0.99)
plt.xticks(np.arange(3, 15+3, step=3))
plt.yticks(np.arange(0.8, 1.02, step=0.1))
plt.tick_params(axis='both', which='major', labelsize=15)
plt.xlabel('Number of negative samples',fontsize=16)
plt.ylabel('Average Precision',fontsize=16)
plt.plot(neg_samples, arr_c1,c='r',marker='x',lw=2, label="c=1")
plt.plot(neg_samples, arr_c2,c='g',marker='x',lw=2, label="c=2")
plt.plot(neg_samples, arr_c3,c='b',marker='x',lw=2, label="c=3")
plt.legend(loc='right',fontsize=14)
plt.tight_layout()
plt.savefig('VaryNegSamples_1_chnImp.png')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
df = pd.DataFrame(columns =['Negative Samples','c1','c2','c3'])
```

```python
df['c1'] = arr_c1
df['c2'] = arr_c2
df['c3'] = arr_c3
df['Negative Samples'] = [3,6,9,12,15]
```

```python
df = df.melt(id_vars=['Negative Samples'], value_vars=['c1','c2','c3'])
```

```python
barWidth = 0.2
plt.figure(figsize=[6,4])
plt.xlim([-1,6.5])
plt.ylim([0.7,1.005])
# set height of bar
bars1 =  arr_c1
bars2 =  arr_c2
bars3 =  arr_c3

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='r', width=barWidth, edgecolor='white', label='p=1')
plt.bar(r2, bars2, color='b', width=barWidth, edgecolor='white', label='p=2')
plt.bar(r3, bars3, color='g', width=barWidth, edgecolor='white', label='p=3')
 
# Add xticks on the middle of the group bars
plt.xlabel('# of Negative Samples', fontsize=16)
plt.ylabel('Average Precision', fontsize=16)
plt.xticks([r + barWidth for r in range(len(bars1))], [3,6,9,12,15],fontsize=14 )

plt.yticks(fontsize=14)
# Create legend & Show graphic
plt.legend(loc='best',fontsize=14)
plt.tight_layout()
plt.savefig('VaryNegSamples_2_chnImp.pdf')
plt.show()
```
